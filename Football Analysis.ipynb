### List of matches and goals for English Premier League (data limited to 10 due to the screen size)

```python
detailed_matches = pd.read_sql("""SELECT Match.id, 
                                        Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season, 
                                        stage, 
                                        date,
                                        HT.team_long_name AS  home_team,
                                        AT.team_long_name AS away_team,
                                        home_team_goal, 
                                        away_team_goal                                        
                                FROM Match
                                Left JOIN Country on Country.id = Match.country_id
                                Left JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name = 'England'
                                AND season = '2015/2016'
                                ORDER by date
                                LIMIT 10;""", conn)
detailed_matches
```

### Case where query showing the result if the team has won or lost

```python
manutd = pd.read_sql("""select m.date, t.team_long_name AS 'opponent', 
                        CASE when m.home_team_goal < m.away_team_goal THEN 'ManUtd Win!'
                        when m.home_team_goal > m.away_team_goal then 'ManUtd Lost!'
                        else 'Tie' END As outcome
                                        FROM match m 
                                        LEFT JOIN Team t
                                        ON m.home_team_api_id = t.team_api_id
                                        WHERE m.away_team_api_id = 10260
                        """, conn)
manutd
```

### League that has most no of goals per season

```python
league_by_season = pd.read_sql("""

SELECT
    l.name AS league, 
    avg(m.home_team_goal + m.away_team_goal) AS avg_goals
    FROM league AS L
    LEFT JOIN match AS m
ON l.id = m.country_id

WHERE m.season = '2013/2014'
    GROUP BY league
    """, conn)
league_by_season
```
